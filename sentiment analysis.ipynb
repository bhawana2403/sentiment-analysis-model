```python
import pandas as pd
import re
data = pd.read_csv("assignment_question_data.csv")
```

```python
def username(sentence):
    tweet = re.search('@[^\s]+',sentence)
    return tweet
```

```python
pid = data['pid']
refer_post = []
refer_flag = []
img_flag = []
img = []
url_flag = []
url = []
user_flag = []
user = []
sentiment = []
l = []
for i in data['comment']:
    
    # referred post
    tweet = re.search('/post/\d+',i)
    if tweet is None:
        refer_post.append(False)
        refer_flag.append(False)
    else:
        r = tweet.group()
        refer_post.append(r[6:])
        refer_flag.append(True)
        
    # for url
    tweet = re.search('((www\.[^\s]+)|(https?://[^\s]*/))',i)
    #l = url(i)
    if tweet is None:
        url_flag.append(False)
        url.append(False)
    else:
        url_flag.append(True)
        url.append(tweet.group())
        
    l = username(i)
    if l is None:
        user_flag.append(False)
        user.append(False)
    else:
        name = l.group()
        user_flag.append(True)
        user.append(name[1:])
        
    # For image    
    m = re.findall(r'\d+-*\w+-*\w*-*\w*-*\w*-*\w*\.png|\d+-*\w+-*\w*-*\w*-*\w*-*\w*\.jpg',i)
    #m = img(i)
    if len(m) == 0:
        img_flag.append(False)
        img.append(False)
    else:
       # for j in m:
        img.append(m)
        img_flag.append(True)     
```

```python
dataframe = pd.DataFrame(list(zip(data['pid'],data['comment'],img_flag,img,url_flag,url,refer_flag,refer_post,user_flag,user,data['Sentiment_for_model'])),columns = ['pid','comment','Image_flag','Image_mentioned','url_flag','url_mentioned','Reference_flag','Reference_post','User_mentioned_flag','User_mentioned','Sentiment_for_model'])        
dataframe
```

```python
dataframe.to_excel("output.xlsx",index_label=False,index=False,header=True)
```
