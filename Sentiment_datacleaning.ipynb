```python
import pandas as pd
import re
data = pd.read_csv("assignment_question_data.csv")
data = data[data.Sentiment_for_model != 'Neutral']
data
```

```python
r = []
#def processRow(row):
for tweet in data['comment']:
    #Lower case
    tweet.lower()
    #convert any url to URL
    tweet = re.sub('((www\.[^\s]+)|(https?://[^\s]+))','url',tweet)
    #Convert any @Username to "AT_USER"
    tweet = re.sub('@[^\s]+','at_user',tweet)
    #Remove image reference
    tweet = re.sub(r'\d+-*\w+-*\w*-*\w*-*\w*-*\w*\.png|\d+-*\w+-*\w*-*\w*-*\w*-*\w*\.jpg','img',tweet)
    #Remove post reference
    
    #Remove additional white spaces
    tweet = re.sub('[\s]+', ' ', tweet)
    tweet = re.sub('[\n]+', ' ', tweet)
    #Remove not alphanumeric symbols white spaces
    tweet = re.sub(r'[^\w]', ' ', tweet)
    #Replace #word with word
    tweet = re.sub(r'#([^\s]+)', r'\1', tweet)
    #Remove :( or :)
    tweet = tweet.replace(':)','')
    tweet = tweet.replace(':(','')
    #trim
    tweet = tweet.strip('\'"')
    # remove digits
    tweet = re.sub(r'\d+','',tweet)
    #remove ssame words
    pattern = re.compile(r"(.)\1{1,}", re.DOTALL)
    tweet = pattern.sub(r"\1\1", tweet)
    print(tweet)
    r.append(tweet)
data['comment'] = r

data['comment']
```

```python
from nltk.corpus import stopwords
import string
stops = set(stopwords.words('english'))
punctuations = list(string.punctuation)
l = ['url','at_user','img']
stops.update(l)
stops.update(punctuations)
```

```python
from nltk.corpus import wordnet

def get_simple_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else: 
        return wordnet.NOUN
```

```python
from nltk import pos_tag
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

result = []
for i,sent in zip(data['comment'],data['Sentiment_for_model']):
    l = []
    words = word_tokenize(i)
    c = [w for w in words if not w.lower() in stops]
    pos = pos_tag(c)
    for i in pos:
        clean_word = lemmatizer.lemmatize(i[0], pos = get_simple_pos(i[1]))
        l.append(clean_word)
    # remove two or more repetition of same word
    result.append((l,sent))


print(result)
```

```python
len(result)  
```

```python
training = result[0:255]
test = result[255:]
```

# Building Freature Set

```python
all_words = []
for i in training:
    all_words += i[0]
print(len(all_words))
```

```python
# frequency distribution of all words
import nltk 
freq = nltk.FreqDist(all_words)
most_common = freq.most_common(500)
len(freq)
```

```python
features = [i[0] for i in most_common]
features
```

```python
def get_feature_dict(words):
    current_features = {}
    words_set = set(words)
    for w in features:
        current_features[w] = w in words_set
    return current_features
```

```python
training_data = [(get_feature_dict(comment), sent) for comment, sent in training]
```

```python
test_data = [(get_feature_dict(comment), sent) for comment, sent in test]
```

```python
training_data[0]
```

# classification using NLTK Naive Bayes 

```python
# we create the data feature set in a fomat that naive baiyes classifier require in nltk
from nltk import NaiveBayesClassifier
classifier = NaiveBayesClassifier.train(training_data)
```

```python
nltk.classify.accuracy(classifier, test_data)
```

```python
classifier.show_most_informative_features(15)
```

```python
test_result = []
gold_result = []

for i in range(len(test_data)):
    test_result.append(classifier.classify(test_data[i][0]))
    gold_result.append(test_data[i][1])
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix   

print('\nClasification report:\n', classification_report(gold_result, test_result))
print('\nConfussion matrix:\n',confusion_matrix(gold_result, test_result))    
```
